```python
import dash
from dash import dcc
from dash import html
```

```python
app = dash.Dash()
```

```python
app.layout = html.Div([
    html.H1("Hello Dash"),
    html.Div("Dash - A data product development framework from plotpy"),

    dcc.Graph(
        id = 'sample-graph',
        {'x' : [5,6,7], 'y' : [12,15,18], 'type' : 'bar' , 'name' : 'SF'}
    )
])
```

```python
if __name__ == "__main__":
    app.run_server(port = 4050)
    
     
```
